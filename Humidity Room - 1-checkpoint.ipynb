```python
import pandas as pd
import numpy as np
from matplotlib.pyplot import *
import matplotlib.pylab as plt
import seaborn as sns
plt.close("all")
import matplotlib.ticker as mticker
import matplotlib.dates as mdates
%matplotlib inline
```

```python
df_2 = pd.read_csv('humidity_room1_datetime.csv', parse_dates=["Timestamp"], index_col ="Timestamp", encoding='utf-8')
print(df_2)
```

```python
df_max = df_2.groupby(pd.Grouper(freq='d')).max().dropna().rename(columns={"humidity":"humidity_max"}).copy()
print(df_max)
```

```python
df_min = df_2.groupby(pd.Grouper(freq='d')).min().dropna().rename(columns={"humidity":"humidity_min"}).copy()
print(df_min)
```

```python
df_mean = df_2.groupby(pd.Grouper(freq='d')).mean().dropna().rename(columns={"humidity":"humidity_mean"}).copy()
print(df_mean)

###SAVE File for Tiago ###

#df_mean.to_csv('room1_humidity_june_2021.csv')
```

```python
frames = [df_max, df_min, df_mean]

result = pd.concat(frames, axis=1, join="inner").copy()

print(result)
```

```python
df = pd.DataFrame(result, columns=['humidity_max','humidity_min','humidity_mean'])

boxplot = df.boxplot(column=['humidity_max','humidity_min','humidity_mean'],patch_artist=True)
```

```python
result.plot()
plt.grid(color='green',linestyle='--', linewidth=0.5)
#plt.mgrid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,5))
#fig, ax = plt.subplots(1)


ax.set_ylim(0,100)


ax.set_ylabel('Daily Humidity %')
ax.set_xlabel('June 2021')
ax.set_title('June 2021 Room1 Humidity in %')

#ax.plot(result.loc['2021-06-02':'2021-06-24', ['humidity_max','humidity_min','humidity_mean']],marker='o', linestyle='-')
ax.plot(result.loc['2021-06-02':'2021-06-24', ['humidity_max','humidity_mean','humidity_min']],marker='o', linestyle='-')

ax.xaxis.set_major_locator(mdates.DayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d'))


#plt.legend(["humidity_max","humidity_min","humidity_mean"],loc='best')
plt.legend(["Max humidity","Mean humidity","Min humidity"],loc='best')

plt.grid(color='green',linestyle='--', linewidth=0.5)
plt.show()
```
