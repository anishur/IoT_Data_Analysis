```python
import pandas as pd
import numpy as np
from matplotlib.pyplot import *
import matplotlib.pylab as plt
import seaborn as sns
plt.close("all")
import matplotlib.ticker as mticker
import matplotlib.dates as mdates
%matplotlib inline
```

```python
# Room3 Mean Temperature

df_room3_temp = pd.read_csv('temp_room3_datetime.csv', parse_dates=["Timestamp"], index_col ="Timestamp", encoding='utf-8')

df_room3_temp_mean = df_room3_temp.groupby(pd.Grouper(freq='d')).mean().dropna().rename(columns={"temp":"room3_temp_mean"}).copy()

#df.to_csv('file1.csv') # Save the file to .csv

#df_room3_temp_mean.to_csv('room3_temp_mean.csv')

print(df_room3_temp_mean)
```

```python
# Room1 Mean Humidity 


df_room1 = pd.read_csv('humidity_room1_datetime.csv', parse_dates=["Timestamp"], index_col ="Timestamp", encoding='utf-8')
df_mean_room1 = df_room1.groupby(pd.Grouper(freq='d')).mean().dropna().rename(columns={"humidity":"room1_humidity"}).copy()
print(df_mean_room1)
```

```python
# Room2 Mean Humidity 


df_room2 = pd.read_csv('humidity_room2_datetime.csv', parse_dates=["Timestamp"], index_col ="Timestamp", encoding='utf-8')
df_mean_room2 = df_room2.groupby(pd.Grouper(freq='d')).mean().dropna().rename(columns={"humidity":"room2_humidity"}).copy()
print(df_mean_room2)
```

```python
idx = pd.date_range('2021-06-02','2021-06-24')
df_mean_room2.index = pd.DatetimeIndex(df_mean_room2.index)
df_mean_room2 = df_mean_room2.reindex(idx,fill_value=np.nan).copy()
print(df_mean_room2)
```

```python
# Room3 Mean Humidity 


df_room3 = pd.read_csv('humidity_room3_datetime.csv', parse_dates=["Timestamp"], index_col ="Timestamp", encoding='utf-8')
df_mean_room3 = df_room3.groupby(pd.Grouper(freq='d')).mean().dropna().rename(columns={"humidity":"outside_humidity"}).copy()
print(df_mean_room3)
```

```python
idx = pd.date_range('2021-06-02','2021-06-24')
df_mean_room3.index = pd.DatetimeIndex(df_mean_room3.index)
df_mean_room3 = df_mean_room3.reindex(idx,fill_value=np.nan).copy()
print(df_mean_room3)
```

```python
frames = [df_mean_room1,df_mean_room2, df_mean_room3]
result = pd.concat(frames, axis=1, join="inner").copy()

#df.columns = ['Column_title_1','Column_title_2','Column_title_3'] ### to change the columns name
result.columns = ['room1_humidity','room2_humidity','outside_humidity']

print(result)
```

```python
frames_test = [df_mean_room1.loc['2021-06-02':'2021-06-25'],df_mean_room2.loc['2021-06-02':'2021-06-25'], df_mean_room3.loc['2021-06-02':'2021-06-25']]
result = pd.concat(frames_test, axis=1, join="inner").copy()

#df.columns = ['Column_title_1','Column_title_2','Column_title_3'] ### to change the columns name
result.columns = ['room1_humidity','room2_humidity','outside_humidity']

print(result)
```

```python
df = pd.DataFrame(result, columns=['room1_humidity', 'room2_humidity', 'outside_humidity'])
boxplot = df.boxplot(column=['room1_humidity', 'room2_humidity', 'outside_humidity'],patch_artist=True)
boxplot.set_title('June 2021 Room1, Room2 & Outside Humidity in %',fontsize=12)
boxplot.set_ylabel('Humidity in %',fontsize=11)
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.set_ylim(0,100)

ax.plot(df_mean_room1.loc['2021-06-02':'2021-06-24', 'room1_humidity'], marker='o', linestyle='-')
ax.plot(df_mean_room2.loc['2021-06-02':'2021-06-24', 'room2_humidity'], marker='o', linestyle='-')
ax.plot(df_mean_room3.loc['2021-06-02':'2021-06-24', 'outside_humidity'], marker='o', linestyle='--')

ax.set_ylabel('Daily Humidity in %')
ax.set_xlabel('June 2021')
ax.set_title('June 2021 Room1 Room2 and Outside Humidity in %')
#ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.MONDAY))
ax.xaxis.set_major_locator(mdates.DayLocator())
#ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d'))
plt.legend(["Room1 Humidity","Room2 Humidity","Outside Humidity"],loc='best')
plt.grid(True)
plt.show()
```
