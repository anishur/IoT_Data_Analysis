```python
import pandas as pd
import numpy as np
from matplotlib.pyplot import *
import matplotlib.pylab as plt
import seaborn as sns
plt.close("all")
import matplotlib.ticker as mticker
import matplotlib.dates as mdates
%matplotlib inline
```

```python
df_2 = pd.read_csv('temp_room2_datetime.csv', parse_dates=["Timestamp"], index_col="Timestamp", encoding='utf-8')
```

```python
print(df_2.head(50))
```

```python
df_mean = df_2.groupby(pd.Grouper(freq='d')).mean().dropna().rename(columns={"temp":"temp_mean"}).copy()
```

```python
# Save the File to .csv 
#df.to_csv('file1.csv')


print(df_mean)

#df_mean.to_csv('room1_temp_mean.csv')

#SAVE File for Tiago

#df_mean.to_csv('room1_temp_june_2021.csv')
```

```python
df_mean = pd.DataFrame(df_mean, columns=['temp_mean'])

#df_mean.plot.bar()
df_mean.plot.line()
plt.grid()
```

```python
fig, ax = plt.subplots()
ax.plot(df_mean.loc['2021-06-02':'2021-06-24', 'temp_mean'], marker='o', linestyle='-')
ax.set_ylabel('Daily Temp ºC')
ax.set_xlabel('June 2021')
ax.set_title('June 2021 Room2_Temp in ºC')
#ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.MONDAY))
ax.xaxis.set_major_locator(mdates.DayLocator())
#ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d'))
plt.grid(True)
```

```python
points= plt.plot(df_mean.loc['2021-06-02':'2021-06-24', 'temp_mean'], marker='o', linestyle='--')
ax = plt.gca()
plt.autoscale(True, axis='both')
plt.grid(True)
```

```python
#Room2 Temp Max value -- cell-1

df_max_room2 = df_2.groupby(pd.Grouper(freq='d')).max().dropna().rename(columns={"temp":"temp_max"}).copy()
print(df_max_room2)
```

```python
#Room2 Temp Max value -- cell-2

idx = pd.date_range('2021-06-02','2021-06-24')
df_max_room2.index = pd.DatetimeIndex(df_max_room2.index)
df_max_room2 = df_max_room2.reindex(idx,fill_value=np.nan)
print(df_max_room2)
```

```python
#Room2 Temp Max value plot -- cell-3

plt.plot(df_max_room2.loc['2021-06-02':'2021-06-24', 'temp_max'], marker='o', linestyle='-')
ax = plt.gca()
plt.autoscale(True, axis='both')
#plt.grid(True)
ax.set_xlabel('June 2021')
ax.xaxis.set_major_locator(mdates.DayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d'))
plt.grid(True)
```

```python
#Room2 Temp Min value -- cell-1

df_min_room2 = df_2.groupby(pd.Grouper(freq='d')).min().dropna().rename(columns={"temp":"temp_min"}).copy()
print(df_min_room2)
```

```python
#Room2 Temp Min value -- cell-2

idx = pd.date_range('2021-06-02','2021-06-24')
df_min_room2.index = pd.DatetimeIndex(df_min_room2.index)
df_min_room2 = df_min_room2.reindex(idx,fill_value=np.nan)
print(df_min_room2)
```

```python
# Room2 Temp Min - Plot --cell -3 

plt.plot(df_min_room2.loc['2021-06-02':'2021-06-24', 'temp_min'], marker='o', linestyle='-')
ax = plt.gca()
plt.autoscale(True, axis='both')
#plt.grid(True)
ax.set_xlabel('June 2021')
ax.xaxis.set_major_locator(mdates.DayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d'))
plt.grid(True)
```

```python
# Room2 Temp Mean value -- cell -1

df_mean_room2 = df_2.groupby(pd.Grouper(freq='d')).mean().dropna().rename(columns={"temp":"temp_mean"}).copy()
```

```python
# Room2 Temp Mean value -- cell -2

idx = pd.date_range('2021-06-02','2021-06-24')
df_mean_room2.index = pd.DatetimeIndex(df_mean_room2.index)
df_mean_room2 = df_mean_room2.reindex(idx,fill_value=np.nan)
print(df_mean_room2)

#SAVE File for Tiago

#df_mean_room2.to_csv('room2_temp_june_2021.csv')
```

```python
# Room2 Temp Mean - Plot --cell -3 

plt.plot(df_mean_room2.loc['2021-06-02':'2021-06-24', 'temp_mean'], marker='o', linestyle='--')
ax = plt.gca()
plt.autoscale(True, axis='both')
#plt.grid(True)
ax.set_xlabel('June 2021')
ax.xaxis.set_major_locator(mdates.DayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d'))
plt.grid(True)
```

```python
# Room2 Temp Mean - Plot --cell -4 

fig, ax = plt.subplots()
ax.plot(df_mean_room2.loc['2021-06-02':'2021-06-24', 'temp_mean'], marker='o', linestyle='-')
ax.set_ylabel('Daily Temp ºC')
ax.set_xlabel('June 2021')
ax.set_title('June 2021 Room2_Temp in ºC')
#ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.MONDAY))
ax.xaxis.set_major_locator(mdates.DayLocator())
#ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d'))
plt.grid(True)
```

```python
# Room2 temperature [Max, Min , Mean] -- plot -- cell 1

frames = [df_max_room2, df_min_room2, df_mean_room2]

result = pd.concat(frames, axis=1, join="inner").copy()
print(result)

```

```python
# Room2 temperature [Max, Min , Mean] -- Box plot -- cell 2


df = pd.DataFrame(result, columns=['temp_max', 'temp_min', 'temp_mean'])
boxplot = df.boxplot(column=['temp_max', 'temp_min', 'temp_mean'],patch_artist=True)
```

```python
# Room2 temperature [Max, Min , Mean] -- Line plot -- cell 3

result.plot()
plt.grid(color='green',linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Room2 temperature [Max, Min , Mean] -- Line plot -- cell 4

fig, ax = plt.subplots(figsize=(12,5))
#fig, ax = plt.subplots(1)

ax.set_ylim(0,30)

ax.set_ylabel('Daily Temp ºC')
ax.set_xlabel('June 2021')
ax.set_title('June 2021 Room2 Temp in ºC')

ax.plot(result.loc['2021-06-02':'2021-06-24', ['temp_max','temp_mean','temp_min']],marker='o', linestyle='-')

ax.xaxis.set_major_locator(mdates.DayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d'))
plt.legend(["Max Temp","Mean Temp","Min Temp"],loc='best')
plt.grid(color='green',linestyle='--', linewidth=0.5)
plt.show()
```
